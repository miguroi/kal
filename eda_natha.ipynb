### 2.2. Memeriksa Nilai yang Hilang

```python
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_data = pd.DataFrame({
    'Jumlah Nilai Hilang': missing_values,
    'Persentase (%)': missing_percent.round(2)
})

print(missing_data[missing_data['Jumlah Nilai Hilang'] > 0].sort_values('Persentase (%)', ascending=False))
```

Dari analisis nilai yang hilang, kita menemukan beberapa hal penting:

- `wind_gust` memiliki 21.58% nilai yang hilang, yang cukup signifikan dan perlu diperhatikan dalam analisis.
- Kolom lain seperti `visibility`, `country`, dan `name` memiliki persentase nilai hilang yang minimal.

### 2.3. Menangani Nilai yang Hilang

```python
df_clean = df.copy()
```

```python
df_clean = df_clean.drop('state', axis=1)
```

Kolom 'state' memiliki lebih dari 90% nilai yang hilang, sehingga tidak akan memberikan informasi yang berguna dalam analisis. Oleh karena itu, kolom ini dihapus dari dataset untuk menyederhanakan analisis dan menghindari bias karena data yang tidak lengkap.

```python
df_clean['wind_gust_ratio'] = df_clean['wind_gust'] / df_clean['wind_speed']
median_ratio = df_clean['wind_gust_ratio'].median()
df_clean.loc[df_clean['wind_gust'].isnull(), 'wind_gust'] = df_clean.loc[df_clean['wind_gust'].isnull(), 'wind_speed'] * median_ratio
df_clean.drop('wind_gust_ratio', axis=1, inplace=True)
```

Untuk menangani nilai yang hilang pada kolom 'wind_gust' (sekitar 21.58%), pendekatan yang digunakan adalah imputasi berdasarkan hubungan dengan kecepatan angin (wind_speed). Hembusan angin (wind_gust) biasanya memiliki pola yang berkorelasi dengan kecepatan angin. Dengan menghitung rasio median antara wind_gust dan wind_speed, kita dapat mengisi nilai yang hilang dengan perkiraan yang lebih akurat dibandingkan dengan sekadar menggunakan nilai median atau mean.


```python
df_clean['visibility'].fillna(df_clean['visibility'].median(), inplace=True)
df_clean = df_clean.dropna(subset=['country', 'name'])
```

Untuk kolom 'visibility' yang hanya memiliki sekitar 0.61% nilai hilang, imputasi dengan nilai median sudah cukup memadai. Kolom 'country' dan 'name' memiliki sangat sedikit nilai yang hilang (kurang dari 0.05%), sehingga lebih baik menghapus baris-baris tersebut untuk menjaga integritas data.

## 3. Analisis Parameter Cuaca

### 3.1. Visualisasi Distribusi Suhu

```python
plt.figure(figsize=(12, 6))
sns.histplot(df_clean['temperature'], bins=50, kde=True)
plt.title('Distribusi Suhu Global', fontsize=15)
plt.xlabel('Suhu (°C)', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.axvline(df_clean['temperature'].mean(), color='r', linestyle='--', label=f'Rata-rata: {df_clean["temperature"].mean():.2f}°C')
plt.axvline(df_clean['temperature'].median(), color='g', linestyle='--', label=f'Median: {df_clean["temperature"].median():.2f}°C')
plt.legend()
plt.show()
```

Distribusi suhu global menunjukkan pola multimodal dengan beberapa puncak yang berbeda. Suhu rata-rata global adalah 10.23°C dengan median 10.96°C (nilai median sedikit lebih tinggi dari rata-rata, menunjukkan distribusi yang sedikit condong ke kiri). Mayoritas pengamatan terkonsentrasi di sekitar 10-15°C, yang merupakan puncak utama distribusi.

Terlihat juga adanya beberapa puncak lain, seperti di sekitar -5°C, 5°C, dan 25°C. Ini menunjukkan adanya pengelompokan data berdasarkan zona iklim yang berbeda di seluruh dunia. Range suhu dalam dataset cukup luas, dari sekitar -40°C hingga 40°C, mencakup berbagai kondisi iklim dari sangat dingin hingga sangat panas.

### 3.2. Distribusi Suhu Berdasarkan Jenis Cuaca

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='weather_main', y='temperature', data=df_clean)
plt.title('Distribusi Suhu Berdasarkan Jenis Cuaca', fontsize=15)
plt.xlabel('Jenis Cuaca', fontsize=12)
plt.ylabel('Suhu (°C)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Suhu terendah secara konsisten terlihat pada kondisi "Snow" (salju) dengan median sekitar -5°C, yang sesuai dengan ekspektasi karena salju terbentuk pada suhu rendah.

Suhu tertinggi diamati pada kondisi "Sand" (pasir), "Dust" (debu), dan "Smoke" (asap) dengan median sekitar 20-30°C. Ini masuk akal karena kondisi-kondisi tersebut sering terjadi di daerah kering dan panas.

"Clouds" (awan) menunjukkan rentang suhu terluas, dari sekitar -40°C hingga 45°C, menunjukkan bahwa awan dapat terbentuk di hampir semua kondisi suhu.

Beberapa jenis cuaca seperti "Thunderstorm" (badai petir) dan "Haze" (kabut) juga cenderung terjadi pada suhu yang relatif tinggi.

### 3.3. Korelasi Antar Parameter Cuaca

```python
numeric_cols = ['temperature', 'pressure', 'humidity', 'temp_min', 'temp_max',
                'visibility', 'wind_speed', 'wind_deg', 'wind_gust', 'clouds_all',
                'rain_1h', 'snow_1h']

corr_matrix = df_clean[numeric_cols].corr()

plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, square=True)
plt.title('Matriks Korelasi Parameter Cuaca', fontsize=15)
plt.tight_layout()
plt.show()
```

Korelasi terkuat terlihat antara suhu (temperature) dengan suhu minimum (temp_min) dan suhu maksimum (temp_max), mendekati 1.0. Ini adalah hubungan yang diharapkan karena parameter-parameter tersebut mengukur fenomena yang sama.

Tekanan (pressure) memiliki korelasi negatif yang cukup kuat dengan suhu (-0.7), menunjukkan bahwa umumnya tekanan udara cenderung lebih rendah di daerah bersuhu tinggi dan sebaliknya.

Kelembaban (humidity) juga menunjukkan korelasi negatif moderat dengan suhu (-0.25), yang konsisten dengan konsep fisika atmosfer di mana udara panas dapat menampung lebih banyak uap air sebelum mencapai titik jenuh.

Curah salju per jam (snow_1h) berkorelasi negatif dengan suhu, yang logis karena salju hanya terjadi pada suhu rendah.

Visibilitas (visibility), curah hujan per jam (rain_1h), dan kecepatan angin (wind_speed) menunjukkan korelasi positif lemah dengan suhu.

Tutupan awan (clouds_all), hembusan angin (wind_gust), dan arah angin (wind_deg) hampir tidak memiliki korelasi dengan suhu, menunjukkan bahwa parameter-parameter tersebut dipengaruhi oleh faktor-faktor lain selain suhu.

### 3.4. Hubungan antara Kelembaban dan Kondisi Cuaca

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='weather_main', y='humidity', data=df_clean)
plt.title('Distribusi Kelembaban berdasarkan Jenis Cuaca', fontsize=15)
plt.xlabel('Jenis Cuaca', fontsize=12)
plt.ylabel('Kelembaban (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- Fog (Kabut) memiliki kelembaban tertinggi dengan median sekitar 90%, sesuai dengan karakteristik kabut yang terbentuk pada kondisi udara jenuh uap air.
- Sand (Pasir) dan Dust (Debu) memiliki kelembaban terendah (sekitar 20-25%), mencerminkan kondisi kering di lingkungan gurun dan area berdebu.
- Snow (Salju), Mist (Kabut tipis), dan Rain (Hujan) menunjukkan kelembaban tinggi, yang konsisten dengan kebutuhan kelembaban atmosfer untuk pembentukan presipitasi.
- Thunderstorm (Badai petir) menunjukkan rentang kelembaban yang luas, dengan median sekitar 50%, mengindikasikan bahwa badai petir dapat terjadi pada berbagai kondisi kelembaban.
- Clear (Cerah) menunjukkan variasi kelembaban yang signifikan, yang menandakan bahwa langit cerah dapat terjadi pada berbagai tingkat kelembaban udara.

## 4. Analisis Geografis dan Temporal

### 4.1. Analisis Suhu Berdasarkan Negara

```python
# Mengambil 10 negara dengan data terbanyak
top_countries = df_clean['country'].value_counts().head(10).index.tolist()

# Membuat dataset yang hanya berisi negara-negara tersebut
df_top_countries = df_clean[df_clean['country'].isin(top_countries)]

# Menghitung statistik suhu untuk setiap negara
temp_by_country = df_top_countries.groupby('country')['temperature'].agg(['mean', 'median', 'min', 'max', 'std']).round(2)
print(temp_by_country.sort_values('mean', ascending=False))
```

```python
# Visualisasi suhu rata-rata di 10 negara teratas
plt.figure(figsize=(14, 6))
sns.barplot(x=temp_by_country.index, y=temp_by_country['mean'], palette='viridis')
plt.title('Suhu Rata-rata di 10 Negara dengan Data Terbanyak', fontsize=15)
plt.xlabel('Negara', fontsize=12)
plt.ylabel('Suhu Rata-rata (°C)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- Filipina (PH) dan Indonesia (ID) memiliki suhu rata-rata tertinggi (sekitar 24°C), yang konsisten dengan lokasi geografis mereka di kawasan tropis.
- Rusia (RU) dan Amerika Serikat (US) menunjukkan suhu rata-rata terendah (di bawah 0°C).
- Australia (AU) memiliki suhu rata-rata tinggi (22.52°C) dengan variasi yang cukup besar (std 4.86).
- Negara-negara Eropa (DE, FR, IT, ES) menunjukkan pola suhu moderat, berkisar antara 10-12°C.

```python
# Visualisasi distribusi suhu dengan boxplot
plt.figure(figsize=(14, 8))
sns.boxplot(x='country', y='temperature', data=df_top_countries)
plt.title('Distribusi Suhu di 10 Negara dengan Data Terbanyak', fontsize=15)
plt.xlabel('Negara', fontsize=12)
plt.ylabel('Suhu (°C)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Membuat peta dunia dengan scatter plot
plt.figure(figsize=(16, 8))

# Membuat scatter plot dengan longitude (x) dan latitude (y)
scatter = plt.scatter(df_clean['lon'], df_clean['lat'],
                      c=df_clean['temperature'], # Warna berdasarkan suhu
                      cmap='coolwarm',
                      alpha=0.5,
                      s=5)  # Ukuran titik

# Menyesuaikan batas peta
plt.xlim(-180, 180)
plt.ylim(-90, 90)

# Menambahkan garis grid untuk menunjukkan koordinat
plt.grid(True, linestyle='--', alpha=0.5)

# Menambahkan judul dan label
plt.title('Distribusi Geografis Data Cuaca Global', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

# Menambahkan colorbar untuk menunjukkan skala suhu
cbar = plt.colorbar(scatter)
cbar.set_label('Suhu (°C)', fontsize=12)

# Tampilkan plot
plt.tight_layout()
plt.show()
```

- Negara-negara tropis (ID, PH) memiliki distribusi suhu yang sempit, menunjukkan kondisi suhu yang relatif stabil sepanjang tahun.
- Negara-negara dengan variasi iklim besar seperti Rusia, China, dan AS memiliki rentang suhu yang sangat luas, dengan outlier yang mencapai -40°C di Rusia.
- Australia menunjukkan distribusi yang menarik dengan rentang yang luas namun condong ke suhu tinggi.
- Peta suhu global menunjukkan gradien yang jelas mengikuti garis lintang (latitude), dengan suhu dingin (biru) di daerah kutub dan suhu panas (merah/coklat) di daerah ekuator.
- Terlihat konsentrasi data yang lebih tinggi di Eropa, Amerika Utara, dan Asia Timur, yang menunjukkan perbedaan infrastruktur pemantauan cuaca antar wilayah.
- Afrika dan Timur Tengah menunjukkan pola suhu yang konsisten tinggi, mencerminkan karakteristik iklim kering dan panas.
- Daerah pegunungan seperti di Asia Tengah dan Amerika Selatan teridentifikasi dari pola suhu yang lebih rendah dibandingkan area sekitarnya pada garis lintang yang sama.

### 4.2. Distribusi Jenis Cuaca di Berbagai Negara

```python
# Membuat tabel kontingensi untuk semua jenis cuaca di 10 negara teratas
weather_country = pd.crosstab(
    df_clean[df_clean['country'].isin(top_countries)]['country'],
    df_clean['weather_main']
)

# Menormalisasi (baris menjadi persentase)
weather_country_pct = weather_country.div(weather_country.sum(axis=1), axis=0) * 100

# Menampilkan tabel persentase
print(weather_country_pct.round(2))
```

```python
# Visualisasi menggunakan heatmap
plt.figure(figsize=(18, 10))
sns.heatmap(weather_country_pct, annot=True, fmt='.1f', cmap='YlGnBu',
            linewidths=0.5, vmin=0, vmax=80)  # Setting vmax ke 80 untuk visualisasi yang lebih baik
plt.title('Distribusi Semua Jenis Cuaca (%) di 10 Negara Teratas', fontsize=15)
plt.xlabel('Jenis Cuaca', fontsize=12)
plt.ylabel('Negara', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Membuat scatter plot berdasarkan jenis cuaca
plt.figure(figsize=(16, 8))

# Membuat dictionary untuk memetakan nilai weather_main ke angka
weather_types = df_clean['weather_main'].unique()
weather_dict = {weather: i for i, weather in enumerate(weather_types)}

# Konversi jenis cuaca ke angka untuk warna
weather_colors = df_clean['weather_main'].map(weather_dict)

# Membuat scatter plot
scatter = plt.scatter(df_clean['lon'], df_clean['lat'],
                      c=weather_colors,
                      cmap='tab20',
                      alpha=0.5,
                      s=5)

# Menyesuaikan batas peta
plt.xlim(-180, 180)
plt.ylim(-90, 90)

# Menambahkan garis grid
plt.grid(True, linestyle='--', alpha=0.5)

# Menambahkan judul dan label
plt.title('Distribusi Geografis Jenis Cuaca Global', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)

# Membuat legend untuk jenis cuaca
legend_elements = [plt.Line2D([0], [0], marker='o', color='w',
                              markerfacecolor=plt.cm.tab20(weather_dict[weather]/len(weather_dict)),
                              markersize=8, label=weather)
                   for weather in weather_types]
plt.legend(handles=legend_elements, title='Jenis Cuaca',
           loc='lower left', bbox_to_anchor=(1.01, 0))

plt.tight_layout()
plt.show()
```

- Awan (Clouds) mendominasi hampir semua negara, dengan persentase tertinggi di Filipina (93.61%) dan Spanyol (80.95%).
- Indonesia (ID) memiliki pola distribusi unik, dengan hampir seimbang antara Clouds (49.23%) dan Rain (50.77%), mencerminkan iklim tropisnya dengan curah hujan tinggi.
- Amerika Serikat (US) memiliki distribusi paling seimbang antara Clear (47.86%) dan Clouds (47.86%).
- Negara-negara dengan iklim dingin seperti Rusia (RU) dan China (CN) menunjukkan persentase Snow yang lebih tinggi dibandingkan negara lain.
- Perancis (FR) memiliki tingkat Rain yang cukup tinggi (35.59%) dibandingkan negara-negara Eropa lainnya.
- Visualisasi peta menunjukkan bahwa kondisi Clear (merah) banyak terkonsentrasi di Amerika Utara, bagian Eropa Selatan, dan Australia bagian tengah.
- Hujan (Rain) terlihat dominan di daerah ekuator, terutama di kawasan hutan hujan Amazon, Afrika Tengah, dan Asia Tenggara.
- Salju (Snow) terkonsentrasi di lintang tinggi utara, terutama di Rusia Utara, Kanada, dan Skandinavia.
- Terlihat pola zona transisi di mana beberapa jenis cuaca bercampur, seperti di perbatasan antara iklim kontinental dan maritim di Eropa dan Asia.
- Daerah gurun seperti Afrika Utara dan Timur Tengah menunjukkan dominasi kondisi Clear, dengan sedikit variasi jenis cuaca lainnya.

- Suhu terendah (-41.21°C) tercatat di zona waktu GMT+11, sementara suhu tertinggi (44.16°C) tercatat di GMT+1.

- Kelembaban berkorelasi negatif dengan suhu (-0.25) - udara hangat bisa menampung lebih banyak uap air.

- Curah hujan tertinggi terjadi pada suhu 15-25°C, terutama di negara tropis.
- Hampir tidak ada hujan signifikan pada suhu ekstrem (di bawah -5°C atau di atas 35°C).

### Pola Geografis dan Waktu
- Konsentrasi data tertinggi terdapat di Eropa, Amerika Utara, dan Asia Timur.
- Zona waktu GMT-3 (Amerika Selatan) memiliki suhu rata-rata tertinggi (27.77°C).
- Zona waktu Amerika Utara (GMT-5 dan GMT-6) memiliki suhu rata-rata terendah (di bawah 1°C).
- Variabilitas suhu tertinggi ada di zona GMT+8 (China, Malaysia) dengan standar deviasi 12.91°C.
- Pola distribusi data mengikuti garis pantai dan pusat populasi, dengan kepadatan rendah di lautan dan daerah terpencil.
