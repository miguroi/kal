# 6. Analisis Multivariat

## 6.1 Korelasi antar variabel numerik

```python
# Pilih kolom numerik
numeric_columns = ['temperature', 'pressure', 'humidity', 'visibility', 'wind_speed', 'clouds_all']
correlation = df[numeric_columns].corr()

# Visualisasi matriks korelasi
plt.figure(figsize=(12, 10))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriks Korelasi Antar Variabel Numerik', fontsize=16)
plt.show()
```

# 8. Analisis Kondisi Cuaca berdasarkan Zona Waktu

```python
# Konversi timezone dari detik ke jam
df['timezone_hours'] = df['timezone'] / 3600

# Kelompokkan berdasarkan zona waktu
timezone_weather = df.groupby(['timezone_hours', 'weather_main']).size().unstack().fillna(0)

# Normalisasi untuk persentase
timezone_weather_percent = timezone_weather.div(timezone_weather.sum(axis=1), axis=0) * 100

# Visualisasi dengan peta panas
plt.figure(figsize=(16, 10))
sns.heatmap(timezone_weather_percent, cmap='YlGnBu', annot=False, fmt='.1f')
plt.title('Distribusi Kondisi Cuaca berdasarkan Zona Waktu (GMT)', fontsize=16)
plt.xlabel('Kondisi Cuaca', fontsize=12)
plt.ylabel('Zona Waktu (GMT+jam)', fontsize=12)
plt.show()
```
